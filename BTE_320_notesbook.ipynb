```python
def get_name():
  user_input = input("What is your name? ")
  if isinstance(user_input, str):
    print(f"Hello, {user_input}!")
  else:
    print("That's not a string!")

get_name()
```

In Python, the `input()` function *always* returns a string, so the `isinstance(user_input, str)` check for `input()` is technically always true. However, this is the correct way to check if any variable holds a string value.
